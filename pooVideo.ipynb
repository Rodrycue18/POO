```python
#ABSTRACCION
class Personaje: #crear clase y los atributos de la clase se deben declarar dentro de init
    def __init__(self,nombre,fuerza,defensa,vida) -> None: #init es constructor donde es una funcion donde recibe los atributos de mi clase
        #self hace referencia asi mismo y accede a atributos y metodos
        #osea self es mi llave a la clase y cada que escribo self logra cambiar algun atributo o metodo de mi clase
        self.nombre = nombre
        self.fuerza = fuerza
        self.defensa=defensa
        self.vida=vida
    
    #creamos un metodo o funcion que nos muestre el estado de los atributos de la clase
    def atributos(self):
        #solo necesitamos self que es la llave para acceder a los atributos
        print(f"{self.nombre} ")
        print("* fuerza:",self.fuerza)
        print("* defensa:",self.defensa)
        print("* vida:",self.vida)
    
    def subirNivel(self,fuerza,defensa):
        self.fuerza = self.fuerza + fuerza
        self.defensa = self.defensa + defensa

    def estaVivo(self):
        return self.vida > 0
    
    def morir(self):
        self.vida = 0
        print(f"{self.nombre} a muerto")

    def dano (self,enemigo): #con Enemigo le estamos pasando los atributos de otra instancia personaje
        return self.fuerza - enemigo.defensa
    
    def atacar(self,enemigo):
        dano = self.dano(enemigo) #llamamos la funcion 
        enemigo.vida = enemigo.vida - dano
        if enemigo.estaVivo(): 
            print(f" {self.nombre} ha realizado {dano} puntos de dano a {enemigo.nombre} \n La vida de {enemigo.nombre} es de {enemigo.vida}")
        else:
            enemigo.morir()
miPersonaje = Personaje("Pepe",22,2,100) # Creamos constructor y una vez creado el metodo init debemos pasar los parametros que init requiere
miEnemigo = Personaje("rodri",10,5,20)
# print(miPersonaje.nombre)# Hasta el momento parece como un diccionario
# miPersonaje.atributos()
# miPersonaje.subirNivel(10,20)
# miPersonaje.atributos()
# print(miPersonaje.vida)
# print(miPersonaje.estaVivo())
# print(miPersonaje.dano(miEnemigo)) 
# miPersonaje.atacar(miEnemigo)
# miEnemigo.atributos()
#Encapsular hace que no se puedan acceder a mis atributos y metodos por fuera de la clase
#-------------------------------------------------------->#
#HERENCIA
#AHORA QUEREMOS CREAR NUESTRA CLASE GUERRERO Y QUE TENGA LOS ATRIBUTOS Y METODOS DE PERSONAJE 
class Guerrero(Personaje): #Dentro del Parentesis le decimos a mi clase Guerrero que herede todo de Personaje
    def __init__(self, nombre, fuerza, defensa, vida, espada) -> None: #Queremos que guerrero tenga mas dano de espada , osea agregar mas parametros en el init, pero ahora crearemos uno en Guerrero
        super().__init__(nombre, fuerza, defensa, vida) #Con super() lo que hacemos es traer el constructor de nuesta clase "Padre" de la que traemos las cosas
        self.espada = espada
    def atributos(self):
        super().atributos()
        print("* Espada: ", self.espada)
    def dano(self, enemigo):
        return self.fuerza * self.espada - enemigo.defensa

#POLIMORFISMO CREO UNA FUNCION """POLIMORFICA""" porque recibe cualquier clase mientras tenga los respectivos metodos todo estara bien
def combate(jugador_1, jugador_2):
    turno = 0
    while jugador_1.estaVivo() and jugador_2.estaVivo():
        print("\nTurno", turno)
        print(">>> Acción de ", jugador_1.nombre,":", sep="")
        jugador_1.atacar(jugador_2) # POLIMORFISMO AQUI
        print(">>> Acción de ", jugador_2.nombre,":", sep="")
        jugador_2.atacar(jugador_1) # POLIMORFISMO AQUI
        turno = turno + 1

    if jugador_1.estaVivo():
        print("\nHa ganado", jugador_1.nombre)
    elif jugador_2.estaVivo():
        print("\nHa ganado", jugador_2.nombre)
    else:
        print("\nEmpate")    
guts = Guerrero("guts",10,13,100,3) #En python existe la herencia multiple , una clase puede heredar mas de una clase 
combate(guts,miPersonaje)
#guts.atacar(miPersonaje)
#print(guts.atributos())
```

```python
#ENCAPSULACION
#Encapsular hace que no se puedan acceder a mis atributos y metodos por fuera de la clase
#En PYTHON TODOS LOS METODOS Y ATRIBUTOS SON PUBLICOS "SOLO SIMULA QUE TIENE ENCAPSULACION"
class Personaje: #crear clase y los atributos de la clase se deben declarar dentro de init
    def __init__(self,nombre,fuerza,defensa,vida) -> None: #init es constructor donde es una funcion donde recibe los atributos de mi clase
        #self hace referencia asi mismo y accede a atributos y metodos
        #osea self es mi llave a la clase y cada que escribo self logra cambiar algun atributo o metodo de mi clase
        self.__nombre = nombre #Al agragar __ en los atributos hacemos que sea privados y no podemos cambiar nada de los atributos por fuera de la clase
        self.__fuerza = fuerza
        self.__defensa=defensa
        self.__vida=vida
    
    #creamos un metodo o funcion que nos muestre el estado de los atributos de la clase
    def atributos(self):
        #solo necesitamos self que es la llave para acceder a los atributos
        print(f"{self.__nombre} ")
        print("* fuerza:",self.__fuerza)
        print("* defensa:",self.__defensa)
        print("* vida:",self.__vida)
    
    def subirNivel(self,fuerza,defensa):
        self.__fuerza = self.__fuerza + fuerza
        self.__defensa = self.__defensa + defensa

    def estaVivo(self):
        return self.__vida > 0
    
    def __morir(self):
        self.__vida = 0
        print(f"{self.__nombre} a muerto")

    def dano (self,enemigo): #con Enemigo le estamos pasando los atributos de otra instancia personaje
        return self.__fuerza - enemigo.__defensa
    
    def atacar(self,enemigo):
        dano = self.dano(enemigo) #llamamos la funcion 
        enemigo.__vida = enemigo.__vida - dano
        if enemigo.estaVivo(): 
            print(f" {self.__nombre} ha realizado {dano} puntos de dano a {enemigo.__nombre} \n La vida de {enemigo.__nombre} es de {enemigo.__vida}")
        else:
            enemigo.__morir()
    # PARA ACCEDER Y/O MODIFICAR COSAS PRIVADAS DEBEMOS CREAR FUNCIONES DE GET Y SET
    
    def get_fuerza(self):
        return self.__fuerza
    def set_fuerza(self,fuerza):
        if fuerza < 0:
            print("Estas tonto no existe fuerza negativ")
        else:
            self.__fuerza = fuerza

myPersonaje = Personaje("pene",100,14,100)
myEnemigo = Personaje("Ale",10,0,100)
#print(myPersonaje.atacar(myEnemigo))
print(myPersonaje.get_fuerza())
print(myPersonaje.set_fuerza(5))
print(myPersonaje.get_fuerza())
```

```python
class Conjunto:
    def __init__(self,conj = []) -> None:
        self.conj = conj

    def elemetos(self):
        print(self.conj)

    def conjVacio(self):
        self.conj = []
        return self.conj
    
    def pertenece(self, e):
        return e in self.conj

miConj = Conjunto([1,23,2])
print(miConj.pertenece(4))
    
```
